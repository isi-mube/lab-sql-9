- Create a Python connection with SQL database and retrieve the results of the last two queries (also mentioned below) as dataframes:
  - Check the number of rentals for each customer for May
  - Check the number of rentals for each customer for June

**Hint**: You can store the results from the two queries in two separate dataframes.

```python
from sqlalchemy import create_engine # let's connect to SQL
import pandas as pd
from getpass import getpass # let's connect to SQL
password = getpass("Please, kindly insert your password:")
```

```python
# Creating a safe connection
connection_string = 'mysql+pymysql://root:' + password + '@localhost/sakila'
engine = create_engine(connection_string)
```

```python
data1 = pd.read_sql_query('SELECT * FROM sakila.rentals_may', engine)
data1.sample(5)
```

```python
data2 = pd.read_sql_query('SELECT * FROM sakila.rentals_june', engine)
data2.sample(5)
```

```python
data4 = pd.read_sql_query('SELECT customer_id, COUNT(customer_id) AS number_of_rentals_per_costumer_may FROM rentals_may GROUP BY customer_id', engine)
data4.head(5)
```

```python
data5 = pd.read_sql_query('SELECT customer_id, COUNT(customer_id) AS number_of_rentals_per_costumer_june FROM rentals_june GROUP BY customer_id', engine)
data5.head(5)
```

- Write a function that checks if customer borrowed more or less films in the month of June as compared to May.

**Hint**: For this part, you can create a join between the two dataframes created before, using the merge function available for pandas dataframes. Here is a link to the documentation for the [merge function](https://pandas.pydata.org/pandas-docs/stable/reference/api/pandas.DataFrame.merge.html).

```python
# Merging the two dataframes
data6 = pd.DataFrame.merge(data4,data5)
data6.sample(5)
```

```python
def may_vs_june():
    data6['greater_in_may'] = data6['number_of_rentals_per_costumer_may'] > data6['number_of_rentals_per_costumer_june']
    data6['greater_in_june'] = data6['number_of_rentals_per_costumer_may'] < data6['number_of_rentals_per_costumer_june']
    return data6
```
